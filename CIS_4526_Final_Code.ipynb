```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
train_data_frame = pd.read_csv('final_train.csv')
dev_data_frame = pd.read_csv('final_dev.csv')
test_data_frame = pd.read_csv('final_test.csv')
scaler = StandardScaler()
y_train_val = train_data_frame["ground_truth"]
x_train_val = scaler.fit_transform(train_data_frame.drop('ground_truth', axis=1))
y_dev_val = dev_data_frame["ground_truth"]
x_dev_val = scaler.transform(dev_data_frame.drop('ground_truth', axis=1))
#Dummy values for tests are not used in the prediction
y_test_val = test_data_frame["ground_truth"]
x_test_val = scaler.transform(test_data_frame.drop('ground_truth',axis=1))
```

```python
# Shift to putting data in Pytorch
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import numpy as np
from sklearn.metrics import f1_score, precision_score, recall_score
```

```python
# Put Data into Dataset so it can be used by DataLoader 
class BinaryClassificationDataset(Dataset):
    def __init__(self,input_arr,target_arr):
        self.input_arr = input_arr
        self.target_arr = target_arr
    def __len__(self):
        return len(self.input_arr)
    #GetItem Method recieves an index and returns the same point for index everytime and converts it into a tensor
    def __getitem__(self,index):
        x = torch.tensor(self.input_arr[index],dtype=torch.float)
        y = torch.tensor(self.target_arr[index],dtype=torch.float)
        return x,y
        
```

```python
#Create network itself 
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(12,12),
            nn.Dropout(p=0.25),
            nn.ReLU(),
            nn.Linear(12,12),
            nn.Dropout(p=0.25),
            nn.ReLU(),
            nn.Linear(12,1),
        )

    def forward(self, x):
        return self.linear_relu_stack(x)
```

```python
BATCH_SIZE = 12
device = "cpu"
```

```python
# Create DataSet and DataLoaders 
train_dataset = BinaryClassificationDataset(x_train_val,np.array(y_train_val))
validation_dataset = BinaryClassificationDataset(x_dev_val,np.array(y_dev_val))
train_dataloader = DataLoader(train_dataset,batch_size=BATCH_SIZE,shuffle=True)
val_dataloader = DataLoader(validation_dataset,batch_size=BATCH_SIZE,shuffle=False)
test_dataset = BinaryClassificationDataset(x_test_val,np.array(y_test_val))
test_dataloader = DataLoader(test_dataset,batch_size=BATCH_SIZE,shuffle=False)

```

```python
model = MLP().to(device)
print(model)
# This loss function does the sigmoid function implicitly
loss_function = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
def train_epoch(dataloader,model,loss_function,optimizer):
    #Model train initates things like dropout
    model.train()
    #Keep track of correct predictions and train losses
    correct = 0 
    train_losses = []
    prediction_vector = list()
    ground_truth_vector = list()
    for batch_index, (X,y) in enumerate(dataloader):
        #Have to unsqueeze to combat mismatch dimension
        X,y = X.to(device), y.to(device).unsqueeze(dim=1)
        pred = model(X)
        loss = loss_function(pred,y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())
        # To make probability between zero or one put it through sigmoid
        correct += (((torch.sigmoid(pred) > 0.5) * 1.0)== y).sum()
        prediction_vector_for_batch_torch_form = ((torch.sigmoid(pred)>0.5) * 1.0)
        prediction_vector_for_batch = prediction_vector_for_batch_torch_form.flatten().tolist()
        ground_truth_vector_for_batch = y.flatten().tolist()
        for value in prediction_vector_for_batch:
            prediction_vector.append(value)
        for value in ground_truth_vector_for_batch:
            ground_truth_vector.append(value)
    print(f"Train Accuracy: {correct/len(dataloader.dataset)}, Train F1-Score: {f1_score(prediction_vector,ground_truth_vector)}")
    return train_losses
def test_epoch(dataloader, model, loss_fn):
    model.eval()
    prediction_vector = list()
    ground_truth_vector = list()
    test_loss, correct = 0, 0
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device).unsqueeze(dim=1)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (((torch.sigmoid(pred) > 0.5) * 1.0) == y).sum()
            prediction_vector_for_batch_torch_form = ((torch.sigmoid(pred)>0.5) * 1.0)
            prediction_vector_for_batch = prediction_vector_for_batch_torch_form.flatten().tolist()
            ground_truth_vector_for_batch = y.flatten().tolist()
            for value in prediction_vector_for_batch:
                prediction_vector.append(value)
            for value in ground_truth_vector_for_batch:
                ground_truth_vector.append(value)
    print(f"Test Avg Loss: {test_loss/len(dataloader)}, Test Accuracy: {correct/len(dataloader.dataset)},Test F1-Score {f1_score(prediction_vector,ground_truth_vector)}, Test Precision{precision_score(prediction_vector,ground_truth_vector)}, Test Recall {recall_score(prediction_vector,ground_truth_vector)}")
    return test_loss/len(dataloader)
```

```python
EPOCHS = 300
train_losses = []
test_losses = []
epoch_num = 1 
for t in range(EPOCHS):
    print(f"Epoch Nr. {t+1} -----------------------")
    train_losses += train_epoch(train_dataloader, model, loss_function, optimizer)
    test_losses.append(test_epoch(train_dataloader,model,loss_function))
    test_losses.append(test_epoch(val_dataloader, model, loss_function))
    file_name = "epoch_num_rehersal"+str(epoch_num)+"model.pth"
    torch.save(model.state_dict(),file_name)
    epoch_num = epoch_num +1
```

```python
#Select model to and load to generate test file
#Able to load model 
#Then create a generate file function and calculate the f1 score that way  
model_to_load = MLP().to(device)
model_to_load.load_state_dict(torch.load("epoch_num_rehersal97model.pth"))
test_epoch(val_dataloader,model_to_load,loss_function)
```

```python
#generate a file function_for_dev 
def generate_file_for_dev(dataloader, model, file_name):
    model.eval()
    prediction_vector = list()
    with torch.no_grad():
        for X, y in dataloader:
            X = X.to(device)
            pred = model(X)
            prediction_vector_for_batch_torch_form = ((torch.sigmoid(pred)>0.5) * 1.0)
            prediction_vector_for_batch = prediction_vector_for_batch_torch_form.flatten().tolist()
            for value in prediction_vector_for_batch:
                prediction_vector.append(value)
    #Open prediction file to write
    counter = 0 
    with open (file_name,'w',encoding='utf-8') as write_file:
        for value in prediction_vector:
            write_file.write("dev_id_"+str(counter)+"\t"+str(int(value))+"\n")
            counter = counter + 1
```
